# tertinggi harga

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data asli
file_path = "Sport car price.csv"
df = pd.read_csv(file_path)

# Pastikan ada kolom Year, jika tidak ada tambahkan tahun default (misalnya 2021)
if "Year" not in df.columns:
    df["Year"] = 2021  # Atur tahun default jika tidak ada dalam dataset

# Bersihkan harga
df["Price (in USD)"] = df["Price (in USD)"].replace(",", "", regex=True).astype(float)

# Ambil hanya mobil dari tahun 2021
df_2021 = df[df["Year"] == 2021]

# Ambil harga **terendah** per merek di tahun 2021
df_lowest = df_2021.loc[df_2021.groupby("Car Make")["Price (in USD)"].idxmin(), ["Car Make", "Year", "Price (in USD)"]]

# **Urutkan dari harga terendah ke tertinggi**
df_lowest = df_lowest.sort_values(by="Price (in USD)", ascending=True).reset_index(drop=True)

# Buat daftar warna berdasarkan colormap
colors = plt.cm.plasma(np.linspace(0, 1, len(df_lowest)))

# Membuat figure dan mengubah warna background
fig, ax = plt.subplots(figsize=(12, 6))
fig.set_facecolor("#E3F2FD")  # Background seluruh figure
ax.set_facecolor("#F9F9F9")  # Background area plot

# Buat diagram batang berdasarkan harga **terendah**
ax.bar(df_lowest["Car Make"], df_lowest["Price (in USD)"], color=colors)

# Label sumbu X sesuai Car Make
plt.xticks(rotation=45, ha="right", fontsize=9)

# Judul
plt.title("Harga Sport Car Terendah di Tahun 2021", fontsize=16, weight='bold')

# Grid horizontal
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.5)

# Format label sumbu Y dengan pemisah ribuan
ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
plt.ylabel("Harga (USD)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load data asli
file_path = "Sport car price.csv"
df = pd.read_csv(file_path)

# Pastikan ada kolom Year, jika tidak ada tambahkan tahun default (misalnya 2021)
if "Year" not in df.columns:
    df["Year"] = 2021  # Atur tahun default jika tidak ada dalam dataset

# Bersihkan harga
df["Price (in USD)"] = df["Price (in USD)"].replace(",", "", regex=True).astype(float)

# Ambil hanya mobil dari tahun 2021
df_2021 = df[df["Year"] == 2021]

# Ambil harga **terendah** per merek di tahun 2021
df_lowest = df_2021.loc[df_2021.groupby("Car Make")["Price (in USD)"].idxmin(), ["Car Make", "Year", "Price (in USD)"]]

# **Urutkan dari harga terendah ke tertinggi**
df_lowest = df_lowest.sort_values(by="Price (in USD)", ascending=True).reset_index(drop=True)

# **Simpan hasil ke file Excel**
output_file = "Sport_car_price_lowest_2021.xlsx"
df_lowest.to_excel(output_file, sheet_name="Harga Terendah 2021", index=False)

print(f"File berhasil disimpan sebagai {output_file}")
```

# hp 2021

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data asli
file_path = "Sport car price.csv"
df = pd.read_csv(file_path)

# Pastikan ada kolom Year, jika tidak ada tambahkan tahun default (misalnya 2021)
if "Year" not in df.columns:
    df["Year"] = 2021  # Atur tahun default jika tidak ada dalam dataset

# Bersihkan kolom Horsepower dari karakter non-angka
df["Horsepower"] = df["Horsepower"].str.replace(r"[^\d.]", "", regex=True)

# Konversi ke angka (float), tangani error dengan NaN jika tidak valid
df["Horsepower"] = pd.to_numeric(df["Horsepower"], errors="coerce")

# Hapus baris dengan nilai NaN setelah konversi
df.dropna(subset=["Horsepower"], inplace=True)

# Filter hanya mobil dari tahun 2021
df_2021 = df[df["Year"] == 2021]

# Ambil Horsepower terendah per merek di tahun 2021
df_lowest_hp = df_2021.loc[df_2021.groupby("Car Make")["Horsepower"].idxmin(), ["Car Make", "Year", "Horsepower"]]

# Urutkan dari Horsepower terendah ke tertinggi
df_lowest_hp = df_lowest_hp.sort_values(by="Horsepower", ascending=True).reset_index(drop=True)

# Buat daftar warna berdasarkan colormap
colors = plt.cm.plasma(np.linspace(0, 1, len(df_lowest_hp)))

# Membuat figure dan mengubah warna background
fig, ax = plt.subplots(figsize=(12, 6))
fig.set_facecolor("#E3F2FD")  # Background seluruh figure
ax.set_facecolor("#F9F9F9")  # Background area plot

# Buat diagram batang berdasarkan Horsepower terendah
ax.bar(df_lowest_hp["Car Make"], df_lowest_hp["Horsepower"], color=colors)

# Label sumbu X sesuai Car Make
plt.xticks(rotation=45, ha="right", fontsize=9)

# Judul
plt.title("Horsepower Terendah Sport Car per Merek tahun 2021", fontsize=16, weight='bold')

# Grid horizontal
plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.5)

# Label sumbu Y
plt.ylabel("Horsepower")

plt.tight_layout()
plt.show()
```
